```python
import numpy as np
import numpy.random as rd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def f(theta, E):
    return np.cos(theta)*np.sqrt(2*E)
```

```python
def BoxMuller(N):
    out = np.zeros(N)
    for i in range(N):
        theta = rd.uniform(low = 0, high = 2*np.pi, size = 1)
        E = rd.exponential(scale = 1/1, size = 1)
        ## be careful with scale it is 1 over the exponent of e
        X = f(theta, E)
        out[i] = X
    return out

plt.figure()
t = BoxMuller(10**4)
plt.hist(t, bins = 50, density = True)
xx = np.linspace(-3, 3, 100)
plt.plot(xx, stats.norm.pdf(xx))
```
