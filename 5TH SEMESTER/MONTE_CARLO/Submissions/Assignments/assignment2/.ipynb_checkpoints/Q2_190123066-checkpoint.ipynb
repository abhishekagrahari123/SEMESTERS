```python
from matplotlib import pyplot as plt
import numpy as np
import math
import statistics
```

```python
#Question 2
def linearCongruentialMethod(Xo, m, a, b,N):
    arr1 = []
    arr = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + b) % m)
    for i in range(0,N):
        arr.append(arr1[i]/m)
    return arr


def exponentialDistributionRandomNumber(N, theta):
    arr = linearCongruentialMethod(29,2**35, 101, 91, N)
    ans = []
    for i in range(0,N):
        ins = -theta*math.log(arr[i])
        ans.append(ins)
    return ans

N = [10, 100, 1000, 10000, 100000]
for i in range(5):
    print(f'N = {N[i]}')
    ans = exponentialDistributionRandomNumber(N[i],1)
    print("Sample mean ",statistics.mean(ans))
    print("Sample variance ",statistics.variance(ans))

    count, bins_count = np.histogram(ans, bins=1000)
    pdf = count / sum(count)
    cdf = np.cumsum(pdf)

    plt.figure()
    x = np.arange(0,30,0.5)
    y = (1-np.exp(-x))
    plt.plot(x, y)
    plt.plot(bins_count[1:], cdf, label="CDF")
    plt.show()
```
