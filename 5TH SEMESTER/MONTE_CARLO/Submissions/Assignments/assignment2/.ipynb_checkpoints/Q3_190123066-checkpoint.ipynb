```python
from matplotlib import pyplot as plt
import numpy as np
import math
import statistics
```

```python
#Question 3
def linearCongruentialMethod(Xo, m, a, b,N):
    arr1 = []
    arr = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + b) % m)
    for i in range(0,N):
        arr.append(arr1[i]/m)
    return arr

# got uniformly distributed ui
def ArcSin(N):
    arr = linearCongruentialMethod(29,2**35, 101, 91, N)
    ans = []
    for i in range(0,N):
        ins = 0.5-0.5*math.cos(arr[i]*3.14)
        ans.append(ins)
    return ans

N = [10, 100, 1000, 10000, 100000]
for i in range(5):
    print(f'N = {N[i]}')
    ans = ArcSin(N[i])
    if N[i] == 10:
        print(ans)
    print("Sample mean ",statistics.mean(ans))
    print("Sample variance ",statistics.variance(ans))

    count, bins_count = np.histogram(ans, bins=1000)
    pdf = count / sum(count)
    cdf = np.cumsum(pdf)

    plt.figure()
    x = np.arange(0,1,0.05)
    y = (2/np.pi)*(np.arcsin(np.sqrt(x)))
    plt.plot(x, y)
    plt.plot(bins_count[1:], cdf, label="CDF")
    plt.show()
```

```python
print("Actual mean = 0.5")
print("Actual variance = 0.125")
```
