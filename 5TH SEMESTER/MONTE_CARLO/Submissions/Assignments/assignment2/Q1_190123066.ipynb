```python
def linearCongruentialMethod(Xo, m, a, b, N):
    arr1 = []
    arr = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + b) % m)
    for i in range(0,N):
        arr.append(arr1[i]/m)
    return arr


def U_Generator(num):
    u = linearCongruentialMethod(33,pow(2,60),29, 99, 17)
    for i in range(17, num):
        ins = (u[i - 17] - u[i - 5])
        if ins < 0:
            ins = ins + 1
        u.append(ins)
    return u
        

u = U_Generator(1000001)

# Creating histogram
plt.hist(u[0:999])
plt.title('N = 1000')
plt.xlabel('U(i)')
plt.ylabel('frequency')
plt.show()

# Creating histogram
plt.hist(u[0:9999])
plt.title('N = 10000')
plt.xlabel('U(i)')
plt.ylabel('frequency')
plt.show()

# Creating histogram
plt.hist(u[0:99999])
plt.title('N = 100000')
plt.xlabel('U(i)')
plt.ylabel('frequency')
plt.show()

plt.scatter(u[0:999], u[1:1000], marker=".", s=1)
plt.title('N = 1000')
plt.xlabel('U(i)')
plt.ylabel('U(i+1)')
plt.show()

plt.scatter(u[0:9999], u[1:10000], marker=".", s=1)
plt.title('N = 10000')
plt.xlabel('U(i)')
plt.ylabel('U(i+1)')
plt.show()

plt.scatter(u[0:99999], u[1:100000], marker=".", s=1)
plt.title('N = 100000')
plt.xlabel('U(i)')
plt.ylabel('U(i+1)')
plt.show()
```
