```python
# importing required libraries.
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# radical inverse function 
def ϕ(base,i):
    res = 0
    k = 0
    while i > 0:
        d = i % base
        i = i // base
        res += d*(base**(-k-1))
        k += 1
    return res
```

```python
# question 2

# generates random numbers from [0,1]
(a, c, m, x) = (2057, 1345, pow(2, 50), 3245)
def uniform():
    global x
    x = (a * x + c) % m 
    return x/m

# Generating N values from Van der Corput sequence and LCG for N = 100 and 100000
X_van_der_corput = []
X_lcg = []
for N in [100,10000]:
    for i in range(1,N+1):
        X_van_der_corput.append(ϕ(2,i))
        X_lcg.append(uniform())

    # Plotting histograms
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title(f'Van Der Corput sequence for N = {N}')
    plt.hist(X_van_der_corput)
    plt.subplot(1, 2, 2)
    plt.title(f'LCG for N = {N}')
    plt.hist(X_lcg)
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```
