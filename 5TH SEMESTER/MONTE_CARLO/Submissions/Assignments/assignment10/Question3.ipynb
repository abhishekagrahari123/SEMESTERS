```python
# importing required libraries.
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# radical inverse function 
def Ï•(base,i):
    res = 0
    k = 0
    while i > 0:
        d = i % base
        i = i // base
        res += d*(base**(-k-1))
        k += 1
    return res
```

```python
# question 3
# Generated and plotted first N values from the Halton sequence Xi = (Ï•_2 (i),Ï•_3 (i)) for N = 100 and 100000
for N in [100, 100000]:
    X = []
    Y = []
    for i in range(1,N+1):
        X.append(Ï•(2,i))
        Y.append(Ï•(3,i))
    plt.figure(figsize=(6, 6))
    plt.scatter(X,Y)
    plt.title(f'Halton sequence for N = {N}')
    plt.xlabel("Ï•2(i)")
    plt.ylabel("Ï•3(i)")
    plt.show()
```
