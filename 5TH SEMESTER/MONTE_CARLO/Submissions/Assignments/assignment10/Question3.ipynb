```python
# importing required libraries.
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# radical inverse function 
def ϕ(base,i):
    res = 0
    k = 0
    while i > 0:
        d = i % base
        i = i // base
        res += d*(base**(-k-1))
        k += 1
    return res
```

```python
# question 3
# Generated and plotted first N values from the Halton sequence Xi = (ϕ_2 (i),ϕ_3 (i)) for N = 100 and 100000
for N in [100, 100000]:
    X = []
    Y = []
    for i in range(1,N+1):
        X.append(ϕ(2,i))
        Y.append(ϕ(3,i))
    plt.figure(figsize=(6, 6))
    plt.scatter(X,Y)
    plt.title(f'Halton sequence for N = {N}')
    plt.xlabel("ϕ2(i)")
    plt.ylabel("ϕ3(i)")
    plt.show()
```
