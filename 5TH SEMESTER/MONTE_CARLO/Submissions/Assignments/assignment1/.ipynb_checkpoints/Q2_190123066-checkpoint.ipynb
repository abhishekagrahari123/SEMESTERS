```python
import math
import numpy as np
import numpy.random as rd
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
from statistics import mean
from statistics import variance
from scipy.stats import gaussian_kde
import statistics
import seaborn as sns
import time
from tabulate import tabulate
```

```python
def linearCongruentialMethod(Xo, m, a, b,N):
    arr1 = []
    arr = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + b) % m)
    for i in range(0,N):
        arr.append(arr1[i]/m)
    return arr

for i in range(1,6):
    a = 1597
    b = 0
    m = 244944
    N = 10000
    Xo = i;
    arr = linearCongruentialMethod(Xo, m, a, b, N)
    plt.figure()
    plt.title(f'Xo = {Xo}, m = {m}, a = {a}, b = {b}')
    plt.hist(arr,20)
    plt.show()

for i in range(1,6):
    a = 51749
    b = 0
    m = 244944
    N = 10000
    Xo = i;
    arr = linearCongruentialMethod(Xo, m, a, b, N)
    plt.figure()
    plt.title(f'Xo = {Xo}, m = {m}, a = {a}, b = {b}')
    plt.hist(arr,20)
    plt.show()
```

```python
U = []
for x in [1, 19, 28, 551, 77]: 
    (a,c,m) = (1597, 0, 244944)
    U1 = linearCongruentialMethod(x,m,a,c,10000)
    U.append(U1)
plt.figure(figsize=(16,12))
(n, bins, patches) = plt.hist(U, 20, label=[1, 19, 28, 551, 77])
plt.legend()
plt.show()
```

```python
print(tabulate({ "Seed":[1, 19, 28, 551, 77], "Values":n}, headers="keys"))
```
