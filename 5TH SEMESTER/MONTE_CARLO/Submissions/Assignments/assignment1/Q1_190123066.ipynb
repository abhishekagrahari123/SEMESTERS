```python
def linearCongruentialMethod(Xo, m, a, b, N):
    arr1 = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + b) % m)
    return arr1


a = 6
b = 0
m = 11
N = 13
for Xo in range(0,11):
    arr = linearCongruentialMethod(Xo, m, a, b,N)
    print(f"Xo = {Xo}, a = {a}, b = {b}, m = {m}")
    for i in range(N):
        print(arr[i], end = " ")
    print("\n")
    
print('____________________________________________________________')

a = 3
b = 0
m = 11
N = 13
for Xo in range(0,11):
    arr = linearCongruentialMethod(Xo, m, a, b,N)
    print(f"Xo = {Xo}, a = {a}, b = {b}, m = {m}")
    for i in range(N):
        print(arr[i], end = " ")
    print("\n")
```
