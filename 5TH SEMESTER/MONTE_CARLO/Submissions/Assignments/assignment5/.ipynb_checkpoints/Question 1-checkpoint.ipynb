```python
import math
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
# this function generates given no of random numbers from [0,1]
# Xo is the seed
# full period if m is a power of 2, c is odd and a is 4*n + 1
def Uniform_Distribution(Xo, m, a, c,N):
    arr1 = []
    arr = []
    arr1.append(Xo%m)
    for i in range(1, N):
        arr1.append((arr1[i - 1] * a + c) % m)
    for i in range(0,N):
        arr.append(arr1[i]/m)
    return arr
```

```python
# it gives N random numbers from F(x) = 1-expo(-x/theta) distribution
def Exponential_Distribution(N,θ,Xo, m, a, c):
#     generating form uniform [0,1]
    U = Uniform_Distribution(Xo, m, a, c,N)
    expo = []
    for i in range(0,N):
#     applying inverse of CDF on unifom [0,1]
        num = -θ*math.log(U[i])
        expo.append(num)
    return expo
```

```python
def Box_Muller(N, Xo1, m1, a1, c1, Xo2, m2, a2, c2):
# generating random theta between 0 to 2*pi
    U = Uniform_Distribution(Xo1, m1, a1, c1, N)
    random_theta = []
    for i in range(0,N):
        random_theta.append(U[i]*2*np.pi)

# generating exponential of mean 1
    expo = Exponential_Distribution(N, 1, Xo2, m2, a2, c2)

# generating N(0,1) 
    out = np.zeros(N)
    for i in range(N):
        theta = random_theta[i]
        E = expo[i]
        X = np.sin(theta)*np.sqrt(2*E)
        out[i] = X
    return out
```

```python
Z1 = Box_Muller(10**4, 43, 2**59, 21, 7, 34, 2**66, 17, 51)
Z2 = Box_Muller(10**4, 19, 2**53, 65, 13, 87, 2**62, 33, 89)
```

```python
std1 = 1
std2 = 2
mu = [5,8]
```

```python
a = [-0.5, 0, 0.5, 1] #stores various correlation coefficient values
A = [] #A[i] stores all numbers generated for various a values
for i in range(4):
    cor = a[i] #correlation coefficient
    a11 = std1
    a12 = 0
    a21 = cor*std2
    a22 = np.sqrt(1-cor*cor)*std2
    X1 = []
    X2 = []
    for j in range(10**4):
        X1.append(mu[0] + a11*Z1[j])
        X2.append(mu[1] + a21*Z1[j] + a22*Z2[j])
    A.append([X1,X2])
    
#     finding boundaries of curve
    x_min = np.min(A[i][0])
    x_max = np.max(A[i][0])

    y_min = np.min(A[i][1])
    y_max = np.max(A[i][1])

# generating a mesh grid for plotting histogram
    x_bins = np.linspace(x_min, x_max, 30)
    y_bins = np.linspace(y_min, y_max, 30)

# plotting histogram
    plt.hist2d(A[i][0], A[i][1], bins = [x_bins, y_bins], cmap = plt.get_cmap('hot'))
    plt.colorbar()
    plt.title(f'a = {cor}')
    plt.show()
```

```python
# List containing the variance covariance values
a =  [-0.5, 0, 0.5, 1]
cov_val = []
for i in range(3):
    cov_val.append(2*a[i])

# Setting mean of the distribution
μ = np.array([5,8])


# Iterating over different covariance values
for idx, val in enumerate(cov_val):
    # Initializing the covariance matrix
    Σ = np.array([[1, val], [val, 4]])

    # Generating a Gaussian bivariate distribution with given mean and covariance matrix
    distr = multivariate_normal(μ, Σ)

    x = np.linspace(0, 9, num=100)
    y = np.linspace(0, 14, num=100)
    X, Y = np.meshgrid(x,y)
    
    # Generating the density function for each point in the meshgrid
    pdf = np.zeros(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            pdf[i,j] = distr.pdf([X[i,j], Y[i,j]])
            
#     finding boundaries of curve and generating mesh grid
    x_min = np.min(A[idx][0])
    x_max = np.max(A[idx][0])
    y_min = np.min( A[idx][1])
    y_max = np.max( A[idx][1])
    x_bins = np.linspace(x_min, x_max, 30)
    y_bins = np.linspace(y_min, y_max, 30)
    
#     plotting histogram and contour over it
    plt.figure()
    plt.hist2d(A[idx][0], A[idx][1], bins = [x_bins, y_bins], cmap = plt.get_cmap('hot'))
    plt.colorbar()
    plt.contour(X,Y, pdf, cmap='jet')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(f'a = {a[idx]}')
plt.show()
```
